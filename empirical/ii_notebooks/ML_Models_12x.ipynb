```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import os 
import sys

import warnings
warnings.filterwarnings('ignore')
```

```python
append_path= os.path.dirname(os.getcwd())
sys.path.append(append_path)
```

```python
data_path = os.path.join(append_path,"data/train_test")
```

```python
X_train = os.path.join(data_path,"X_train.csv")
X_test = os.path.join(data_path,"X_test.csv")
y_train = os.path.join(data_path,"y_train.csv")
y_test = os.path.join(data_path,"y_test.csv")
```

```python
X_train = pd.read_csv(X_train)
X_test = pd.read_csv(X_test)
y_train = pd.read_csv(y_train)
y_test = pd.read_csv(y_test)
```

# Support Vector Classifier

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)
```

```python
# accuracy_score, confusion_matrix and classification_report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

svc_train_acc = accuracy_score(y_train, svc.predict(X_train))
svc_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy of Support Vector Classifier is : {svc_train_acc}")
print(f"Test accuracy of Support Vector Classifier is : {svc_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Quick check for overfitting due to Greater Than > 5-10% difference in Training & Test  accuracy score:

from sklearn.model_selection import cross_val_score

# Compute cross-validation scores (on training data)
cv_scores = cross_val_score(svc, X_train, y_train, cv=5, scoring='accuracy')

# Print results
print(f"Cross-Validation Scores: {cv_scores}")
print(f"CV Mean Accuracy: {cv_scores.mean():.2f}")
print(f"CV Standard Deviation: {cv_scores.std():.2f}")

# Now train on the full training set and evaluate on test data
svc.fit(X_train, y_train)
test_accuracy = svc.score(X_test, y_test)
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
cv_scores
#how do you tell these 5x scores are for training set vs validation set
#comparing it to the learning curve you can tell its for the validation set, the hold-out fold.
```

* Model trained and evaluated on 5x different folds of the training data
    * The models accuracy varies slightly across folds 72% to 76%
    * This suggests consistent performance across different folds of the data (No Major Instability)
* CV Mean Accuracy score 74%. This is now the baseline expectation for how this model will perform on new, similar data.
* The small standard deviation (Â±1%) and test performance suggest the model is stable and generalizes well.
    * Low Variability: The model is stable and not highly sensitive to how the data is split.
    * High Variability: ex: +/- 10%, this would indicate instability or dependence on specific data subsets (folds)
* Conclusion: No Overfitting. Overfitting would show high training accuracy & low test accuracy, indicating model overfits the training data.


```python
from sklearn.model_selection import learning_curve  
train_sizes, train_scores, val_scores = learning_curve(svc, X_train, y_train, cv=5)  
plt.plot(train_sizes, train_scores.mean(axis=1), label='Train')  
plt.plot(train_sizes, val_scores.mean(axis=1), label='Validation')  
plt.show()
```

```python
# Generate learning curve data
#Learning Curve: Plot shows how a models performance score as the training set sample size increases.
#This shows us the evolution of the 2x scores as the sample size of training observations increases.
#Note, the error is less with smaller training size because EX: We take one single instance from the 
    #training set and use it to estimate a model. The error on the training instance will be 0/Very Low,
    #since it's quite easy to perfectly fit a single data point. The Validation Set sapmle size remains the
    #same through out all iteartions of training, thus error on validation set will be very large compared to
    #one instance example.
#A learning curve shows how error changes as the training set size increases.
#Training set n=60, 195, 330, 465, 600.
    #As we increase the training set size, the model cannot fit perfectly anymore the training set.
    #So the training error becomes larger. However, the model is trained on more data, 
    #so it manages to fit better the validation set. Thus, the validation error decreases.
#Learning curves give us an opportunity to diagnose bias and variance in supervised learning models. 
train_sizes, train_scores, val_scores = learning_curve(
    svc, X_train, y_train, cv=5, scoring='accuracy'
)
#cv=5: Data is split into 5 folds (5x groups within), 1x fold is the hold-out fold & used for validation,
    #5x folds within thus 5x training splits/rounds across

#train_scores: This is the subset used for training & models predictions compared to true labels
#val_score: This is the score on the subset not used in training set, isolated for "grading" (held out fold)

# Calculate means and standard deviations
train_mean = train_scores.mean(axis=1)
train_std = train_scores.std(axis=1)
val_mean = val_scores.mean(axis=1)
val_std = val_scores.std(axis=1)

# Plot learning curve
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, label='Training Score', color='blue')
plt.plot(train_sizes, val_mean, label='Validation Score', color='orange')

# Add shaded areas for standard deviation
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='orange')

# Find the index of the minimum validation score (the "turning point")
min_val_idx = np.argmin(train_mean)
training_size_at_min = train_sizes[min_val_idx]

#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@

# Add vertical line at the turning point
plt.axvline(
    x=training_size_at_min, 
    color='red', 
    linestyle='--', 
    label=f'{training_size_at_min}: Point of\nIntersection'
)

#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@

# Add labels and legend
plt.xlabel('Training Samples', fontsize=12)
plt.ylabel('Percent Correct (Accuracy)', fontsize=12)
plt.title('Learning Curve (SVC)',loc = "left", fontsize=18)
plt.legend(loc='upper right')
plt.grid(False)
plt.show()

#NOTE: The x-axis does NOT represent the validation set sizeâit shows how much data the model was trained on
#The validation set size perfold is always, For cv=5, the validation set size per fold is always len(X_train) // 5.
    #Thus 750 // 5 = 150 in this case.
```

* Big Gap between the scores indicates Overfitting.

# Decision Tree Classifier

```python
# accuracy_score, confusion_matrix and classification_report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

dtc_train_acc = accuracy_score(y_train, dtc.predict(X_train))
dtc_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy of Decision Tree is : {dtc_train_acc}")
print(f"Test accuracy of Decision Tree is : {dtc_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#HYPERPARAMETER Tuning

from sklearn.model_selection import GridSearchCV

grid_params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [3, 5, 7, 10],
    'min_samples_split' : range(2, 10, 1),
    'min_samples_leaf' : range(2, 10, 1)
}

grid_search = GridSearchCV(dtc, grid_params, cv = 5, n_jobs = -1, verbose = 1)
grid_search.fit(X_train, y_train)
```

```python
# best estimator 

dtc = grid_search.best_estimator_

y_pred = dtc.predict(X_test)
```

```python
# accuracy_score, confusion_matrix and classification_report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

dtc_train_acc = accuracy_score(y_train, dtc.predict(X_train))
dtc_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy of Decision Tree is : {dtc_train_acc}")
print(f"Test accuracy of Decision Tree is : {dtc_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
